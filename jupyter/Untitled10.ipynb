```python
import csv
import numpy as np


f= open('/content/age2019_02_noComa_cgName.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더제거

data = list(data) #추가

name = input('인구 구조가 알고 싶은 지역은 이름(읍면동 단위)을 입력해주세요. :')

mn = 1 # 최솟값을 저장할 변수 생성 및 초기화
result_name = '' # 최솟값을 갖는 지역의 이름 저장할 변수 생성 및 초기화
result = 0 # 최솟값을 갖는 지역의 연령대별 인구 비율 저장 할 변수 생성 및 초기화

for row in data :
  if name in row[0] :
    home =np.array(row[3:], dtype = int) / int(row[2])  #배열과 배열끼리 연산이 가능하게 해주기위해

for row in data :
  away = np.array(row[3:], dtype = int)/ int(row[2])
  s = np.sum((home-away)**2) #->음수가 나올수록 더 비슷하다고 잘못나옴->음수를 양수로 변환
  if s < mn and name not in row[0] : # 같은 이름인지 확인
    mn = s
    result_name = row[0]
    result = away # 인구 비율 저장
  

#데이터 시각화
plt.rc('font', family='NanumBarunGothic') 
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.title(name+' 지역의 인구 구조')

plt.plot(home, label = name)
plt.plot(result, label = result_name)
plt.legend()
plt.show()
```
