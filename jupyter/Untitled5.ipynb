```python
import csv
import matplotlib.pyplot as plt
```

# 그래프 스타일 지정 (회색 배경의 격자무늬, 2/4 값을 의미하는 선의 색 변경)

1) 히스토그램을 활용한 최고기온, 최저 기온, 평균기온 분포 확인하기
2) 상자그림을 활용한 최고 기온 vs 서울 최고 기온
3) 월별 최고기온 상자 그림 그리기
4) 특정 달의 일별 최고 기온 상자 그림 그리기


```python
f = open("/content/100.csv", "r", encoding="cp949")
data = csv.reader(f)
header = next(data)
tmp_high = []
tmp_low = []
tmp_avg = []
for i in data:
    if i[4] != "":
        tmp_high.append(float(i[4]))
    if i[3] != "":
        tmp_low.append(float(i[3]))
    if i[2] != "":
        tmp_avg.append(float(i[2]))

# 최고 기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(tmp_high, bins=100, color='r') # bins는 구간 옵션, 6구간으로 실행 되도록 하는 것, 결측치를 보고 싶을 때
plt.hist(tmp_low, bins=100, color='b')
plt.hist(tmp_avg, bins=100, color='g')
# bins를 100으로 설정했지만 이것이 최대치임, 이보다 적으면 잘라서 보기 좋게 보여줌
plt.show()

# day = []
# for i in range(31):
#     day
```

```python
f = open("/content/100.csv", "r", encoding="cp949")
data = csv.reader(f)
header = next(data)
f2 = open("/content/seoul.csv", "r", encoding="cp949")

data_seoul = csv.reader(f2)
header = next(data_seoul)
tmp_high_seoul = []
tmp_high = []

for i in data:
    if i[4] != "":
        tmp_high.append(float(i[4]))

for i in data_seoul:
    if i[4] != "":
        tmp_high_seoul.append(float(i[4]))

# 박스플롯 시각화
plt.figure(dpi=100)
plt.boxplot([tmp_high_seoul, tmp_high])
plt.show()
```

```python
f = open("/content/100.csv", "r", encoding="cp949")
data = csv.reader(f)
header = next(data)

tmp_month_high = []
for i in range(12):
    tmp_month_high.append([])

for i in data:
    if i[4] != "":
        tmp_month_high[int(i[0].split("-")[1])-1].append(float(i[4]))

plt.figure(dpi=200)
plt.boxplot(tmp_month_high)
plt.show()
```

```python
f = open("/content/100.csv", "r", encoding="cp949")
data = csv.reader(f)
header = next(data)

tmp_jan = []
for i in range(31):
    tmp_jan.append([])

for i in data:
    if i[0].split("-")[1] == "01":
        if i[4] != "":
            tmp_jan[int(i[0].split("-")[2])-1].append(float(i[4]))


plt.figure
plt.boxplot(tmp_jan)
plt.show()
```
