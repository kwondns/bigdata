```python
import csv
import matplotlib.pyplot as plt

def get_low_high(file):
    low = [0, 0]
    high = [0, 0]
    diff = [0, 0]
    tmp_high = []
    tmp_low = []
    diff_tmp = []
    
    with open(file, "r", encoding="cp949") as file:
        data = csv.reader(file)
        header = next(data)
        for idx, row in enumerate(data):
            if row[3] == '':
                row[3] = 999
            if row[4] == '':
                row[4] = -999
            row[3], row[4] = float(row[3]), float(row[4])
            if row[3] != 999:
                tmp_high.append(row[4])
            if row[4] != -999:
                tmp_low.append(row[3])
            if row[3] != 999 and row[4] != -999:
                diff_tmp.append(row[4]-row[3])
            if idx == 0:
                low[0], low[1], high[0], high[1] = row[0], row[3], row[0], row[4]
            else:
                if low[1] > row[3]:
                    low[0], low[1] = row[0], row[3]
                if high[1] < row[4]:
                    high[0], high[1] = row[0], row[4]
        print(low, high)   
        fig = plt.figure(dpi=75, figsize=(10,12)) 
        a1 = fig.add_subplot(3, 1 ,1)
        a2 = fig.add_subplot(3, 1, 2)
        a3 = fig.add_subplot(3,1,3)
        
        a1.plot(range(len(tmp_high)),tmp_high, "r")
        a1.set_title("high")
        a2.plot(range(len(tmp_low)), tmp_low, "b")
        a2.set_title("low")
        a3.plot(range(len(diff_tmp)), diff_tmp, "g")
        a3.set_title("diff")
        plt.show()
```

```python
get_low_high("/content/파주.csv")
```
