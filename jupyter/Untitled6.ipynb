```python

import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
```

개인 데이터를 조별로 취합하기
1. 성별 데이터 수평 막대 그래프로 시각화(항아리 그래프) - 내 동네의 데이터
2. 2019년 2월 연령별 인구 구조 -> 2020년 2월 연령별 인구 구조 비교(같이 그리기) </br>
-> 조원들의 수 만큼

```python
#1번

f = open('/content/age_gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)
male = []
female = []
for row in data:
  if '일산서구 대화동' in row[0]:
    for i in range(0, 101):
      male.append(-int(row[i + 3])) # 값들이 음수로 들어간다.
      female.append(int(row[-(i+1)])) # -1 값부터 담는다. -> 100세 이상부터 값이 담긴다.
    break


female.reverse() 
# 0세부터 인구수 나올 수 있도록 데이터 뒤집기



# 시각화 설정
plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')
plt.title('일산서구 대화동 지역의 남녀 성별 인구 분포')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.barh(range(101),male, label='남성')
plt.barh(range(101),female, label='여성')
plt.legend()
plt.show()
```

```python
#2번
f1 = open("/content/age2019_02.csv", "r", encoding="cp949")
f2 = open("/content/age2020_02.csv", "r", encoding="cp949")
data_2019 = csv.reader(f1)
data_2020 = csv.reader(f2)
next(data_2019)
next(data_2020)

st = []
nd = []

for i in data_2019:
    if "일산서구 대화동" in i[0]:
        for j in range(0, 101):
            st.append(int(i[j+3]))

for i in data_2020:
    if "일산서구 대화동" in i[0]:
        for j in range(0, 101):
            nd.append(int(i[j+3]))


plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')
plt.title('일산서구 대화동 지역의 연령별 인구 분포')
plt.plot(st, label='2019년')
plt.plot(nd, label='2020년')
plt.legend()
plt.show()
```

```python
#3번 한양대 
f1 = open("/content/2018.csv", "r", encoding="cp949")
f2 = open("/content/2019.csv", "r", encoding="cp949")
f3 = open("/content/2020.csv", "r", encoding="cp949")
f4 = open("/content/2021.csv", "r", encoding="cp949")
data_2018 = csv.reader(f1)
data_2019 = csv.reader(f2)
data_2020 = csv.reader(f3)
data_2021 = csv.reader(f4)
next(data_2018)
next(data_2019)
next(data_2020)
next(data_2021)

list_18_seoul = []
list_19_seoul = []
list_20_seoul = []
list_21_seoul = []
list_18_erica = []
list_19_erica = []
list_20_erica = []
list_21_erica = []

for i in data_2018:
    if "성동구 사근동" in i[0]:
        for j in range(0, 101):
            list_18_seoul.append(int(i[j+3]))
    if "상록구 사동" in i[0]:
        for j in range(0, 101):
            list_18_erica.append(int(i[j+3]))

for i in data_2019:
    if "성동구 사근동" in i[0]:
        for j in range(0, 101):
            list_19_seoul.append(int(i[j+3]))
    if "상록구 사동" in i[0]:
        for j in range(0, 101):
            list_19_erica.append(int(i[j+3]))

for i in data_2020:
    if "성동구 사근동" in i[0]:
        for j in range(0, 101):
            list_20_seoul.append(int(i[j+3]))
    if "상록구 사동" in i[0]:
        for j in range(0, 101):
            list_20_erica.append(int(i[j+3]))

for i in data_2021:
    if "성동구 사근동" in i[0]:
        for j in range(0, 101):
            list_21_seoul.append(int(i[j+3]))
    if "상록구 사동" in i[0]:
        for j in range(0, 101):
            list_21_erica.append(int(i[j+3]))

plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')
plt.title('성동구 사근동 지역의 연령별 인구 분포')
plt.plot(list_18_seoul, label='2018년', color = "g")
plt.plot(list_19_seoul, label='2019년', color = "r")
plt.plot(list_20_seoul, label='2020년', color = "b")
plt.plot(list_21_seoul, label='2021년', color = "purple")


plt.legend()
plt.show()
```

```python
plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')
plt.title('상록구 사동 지역의 연령별 인구 분포')
plt.plot(list_18_erica, label='2018년', color = "g")
plt.plot(list_19_erica, label='2019년', color = "r")
plt.plot(list_20_erica, label='2020년', color = "b")
plt.plot(list_21_erica, label='2021년', color = "purple")


plt.legend()
plt.show()
```
