```python
import csv
import matplotlib.pyplot as plt

f = open("age_gender.csv", encoding="cp949")
data = csv.reader(f)
result = []
location = ["서울특별시","부산광역시", "대구광역시", "인천광역시", "광주광역시", "대전광역시", "울산광역시", "세종특별자치시"]

#name=input("찾고 싶은 지역의 이름을 알려주세요 :")
for name in location:
    globals()[f"result_{name}"] = []
    for row in data: 
        if name in row[0]:
            for i in range(3, 104):
                globals()[f"result_{name}"].append(int(row[i].replace(",","")) - int(row[i+103].replace(",","")))
            break

fig = plt.figure(dpi=75, figsize=(20,14)) 
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')

for idx, name in enumerate(location):
    globals()[f"plot_{name}"] = fig.add_subplot(2, 4 ,idx+1)
    globals()[f"plot_{name}"].bar(range(101),globals()[f"result_{name}"])
    globals()[f"plot_{name}"].set_ylim([-10000,4000])
    globals()[f"plot_{name}"].set_title(f"{name}")
plt.legend()
plt.show()
```

```python
import csv

f = open('/content/age_gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)
male = []
female = []
name = input("찾고 싶은 지역의 이름을 알려주세요 : ")

for row in data:
  if name in row[0]:
    for i in range(0, 101):
      male.append(int(row[i + 3].replace(",",""))) # 값들이 음수로 들어간다.
      female.append(int(row[-(i+1)].replace(",",""))) # -1 값부터 담는다. -> 100세 이상부터 값이 담긴다.
    break


female.reverse() 
# 0세부터 인구수 나올 수 있도록 데이터 뒤집기



# 시각화 설정
plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')

plt.scatter(male,female)
plt.plot(range(max(male)), range(max(male)), "g")
plt.show()
```

```python
plt.figure(dpi=200)
plt.scatter(male, female, c= range(101), cmap="jet", alpha=0.5)
plt.plot(range(max(male)), range(max(male)), "g")
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import math

f = open("age_gender.csv", encoding="cp949")
data = csv.reader(f)

m = []
f = []
size = []

name = input("찾고 싶은 지역의 이름을 알려주세요 : ")
for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(int(i.replace(",","")))
            size.append(math.sqrt(int(i.replace(",",""))))
        for i in row[106:]:
            f.append(int(i.replace(",","")))
        break

plt.figure(dpi=200)
plt.scatter(m,f, s=size, c=range(101), cmap="jet", alpha=0.5)
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), "g")
plt.show
```
